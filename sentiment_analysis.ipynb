```python
import pandas as pd
from google.colab import files
```

```python
df = pd.read_csv('/content/drive/MyDrive/cleaned_complaints.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df['Complaint Category'] = df['Complaint Category'].str.title().str.strip()
```

```python
df['Date Filed'] = pd.to_datetime(df['Date Filed'], errors='coerce')
df.dropna(subset=['Date Filed'], inplace=True)

df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def classify_sentiment(text):
    score = analyzer.polarity_scores(text)['compound']
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Complaint Category'].apply(classify_sentiment)
df[['Complaint Category', 'Sentiment']].head()
```

```python
df.to_csv("cleaned_complaints.csv", index=False)
files.download("cleaned_complaints.csv")
```
